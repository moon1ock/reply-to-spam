```python
import pandas as pd
```

```python
train = pd.read_csv('/Users/andriylunin/Desktop/P11/voice_forensics_train.csv').drop(columns = ['Id','speaker_id','age'])
test = pd.read_csv('voice2genderage_test_x.csv').drop(columns = ['Id'])
train
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils

# for modeling
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
df = train
df['gender'].replace(('male', 'female'), (0, 1), inplace=True)
df
```

```python
df.gender.value_counts()
```

```python
import numpy as np
Y = df['gender']
X = df.drop(['gender'], axis=1)

print(X.shape)
print(Y.shape)
X = np.array(X)
```

```python
# build a model
model = Sequential()
model.add(Dense(32, input_shape=(X.shape[1],), activation='relu')) # Add an input shape! (features,)
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()

# compile the model
model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])


es = EarlyStopping(monitor='val_accuracy',
                                   mode='max',
                                   patience=15,
                                   restore_best_weights=True)


history = model.fit(X,
                    Y,
                    callbacks=[es],
                    epochs=10,
                    batch_size=10,
                    validation_split=0.2,
                    shuffle=True,
                    verbose=1)
```

```python
x_test = np.array(test)
```

```python
np.round(model.predict(x_test),0)
```
